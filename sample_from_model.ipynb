The model in this Notebook was trained using configuration `config/two_secs_augmented/ex_small.yaml`.
That configuration trains a model on LJSpeech (~13K examples) using 160 epochs on the first two seconds of audio.
The training uses a batch size of 64 and an augmentation factor of 4.

I interrupted the training at 89 epochs as the loss was not improving anymore and there were other experiments I wanted to try.

The results are decent on many validation examples.
Many custom examples do not work well however.
So there is some cherry picking below. I might get around to improving the model more ;).

There is a faint clicking background that is probably caused by the augmentation (small gaussian noise) since
non-augmented models don't have that problem. The augmentation does prevent overfitting for the model tested
here. (Though the slightly overfitted model does not perform much worse than this model.)

```python
import IPython
from audio_encoding import text_to_speech_pipeline
```

```python
model_path = "rem_aug/models/ex_small.model"

# Hardcoded constants at the moment
forward_params = dict(
    max_length = 301, # Corresponds to Encodec lowest fidelity of encoding for 2 seconds of audio plus EOS token
    num_beams = 4,
    do_sample = True,
)
```

```python
pipeline = text_to_speech_pipeline(model_path)
```

```python
audio = pipeline("hello friend", forward_params=forward_params)
IPython.display.Audio(audio["audio"], rate=audio["sampling_rate"])
```

```python
audio = pipeline("I am happy you are here", forward_params=forward_params)
IPython.display.Audio(audio["audio"], rate=audio["sampling_rate"])
```

```python
audio = pipeline("The family would enjoy your visit", forward_params=forward_params)
IPython.display.Audio(audio["audio"], rate=audio["sampling_rate"])
```
