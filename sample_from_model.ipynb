```python
import IPython
from audio_encoding import text_to_speech_pipeline
```

```python
model_path = "rem_aug/models/ex_small.model"

# Hardcoded constants at the moment
forward_params = dict(
    max_length = 301, # Corresponds to Encodec lowest fidelity of encoding for 2 seconds of audio plus EOS token
    num_beams = 4,
    do_sample = True,
)
```

```python
pipeline = text_to_speech_pipeline(model_path)
```

One can produce a lot of examples, in particular those from the v

```python
audio = pipeline("hello friend", forward_params=forward_params)
IPython.display.Audio(audio["audio"], rate=audio["sampling_rate"])
```

```python
audio = pipeline("I am happy you are here", forward_params=forward_params)
IPython.display.Audio(audio["audio"], rate=audio["sampling_rate"])
```
